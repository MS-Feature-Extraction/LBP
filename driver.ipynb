## Driver file to show-case LBP (Local Binary Pattern) functions written in utility.py
These functions are defined in utility
- **LBP**
    - Used to generate LBP(Local Binary Pattern) from image
        - Params list: 
          - img: image or matrix to construct glcm from
          - r: Radius
          - p: sampling points
        - Returns the LBP image/matrix
- Utility functions
    - **extract_neighbourhood**: To extract neighbourhood
        - Params list:
            - I (img)
            - x, y
            - r (radius)
        - Returns array of 4 sides in clockwise direction
    - **extract_pattern**: To extract the pattern (bit-string) from the neighbourhood
        - Params list:
            - neighbourhood: output from **extract_neighbourhood**
            - pixel: value of the center pixel value
            - p: sampling points
